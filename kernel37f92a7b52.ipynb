```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
sns.set(style="ticks")
```

```python
data = pd.read_csv('../input/toy-dataset/toy_dataset.csv', sep=",")
```

```python
data.head()
```

```python
data.shape
```

```python
total_count = data.shape[0]
print('Всего строк: {}'.format(total_count))
```

```python
data.dtypes
```

```python
# Проверим наличие пустых значений
# Цикл по колонкам датасета
for col in data.columns:
    # Количество пустых значений - все значения заполнены
    temp_null_count = data[data[col].isnull()].shape[0]
    print('{} - {}'.format(col, temp_null_count))
```

```python
data.isnull().sum()
```

```python
#Диаграмма рассеяния для City и Income
fig, ax = plt.subplots(figsize=(10,10)) 
sns.scatterplot(ax=ax, x='City', y='Income', data=data)
```

```python
#Гистограмма
fig, ax = plt.subplots(figsize=(10,10)) 
sns.distplot(data['Income'])
```

```python
#Комбинация гистограмм и диаграмм рассеивания
sns.jointplot(x='Age', y='Income', data=data)
```

```python
sns.jointplot(x='Age', y='Income', data=data, kind="hex")
```

```python
sns.jointplot(x='Age', y='Income', data=data, kind="kde")
```

```python
#Парные диаграммы - комбинация гистограмм и диаграмм рассеивания для всего набора данных
sns.pairplot(data)
```

```python
#С помощью параметра "hue" возможна группировка по значениям какого-либо признака
sns.pairplot(data, hue="Illness")
```

```python
#Ящик с усами
sns.boxplot(x=data['Income'])
```

```python
#Ящик с усами
sns.boxplot(y=data['Income'])
```

```python
# Распределение параметра Income сгруппированные по City.
sns.boxplot(x='City', y='Income', data=data)
```

```python
#Violin plot
sns.violinplot(x=data['Income'])
```

```python
# Распределение параметра Income сгруппированные по City.
sns.violinplot(x='City', y='Income', data=data)
```

```python
#Корреляция
data.corr()
```

```python
sns.heatmap(data.corr())
```

```python
# Вывод значений в ячейках
sns.heatmap(data.corr(), annot=True, fmt='.3f')
```

Я построил: диаграмму рассеяния - позволяет обнаружить наличие зависимости, гистограмму - позволяет оценить плотность вероятности распределения данных, jointplot - комбинация гистограмм и диаграмм рассеивания, парные диаграммы - комбинация гистограмм и диаграмм рассеивания для всего набора данных, ящик с усами - отображает одномерное распределение вероятности.

Выводы по диаграммам: наибольший Income - в городе Mountain View, наименьший - Dallas. Наиболее часто встречающийся Income - 100000.

Income отчасти коррелирует с Number, слабо коррелирует с Age.
